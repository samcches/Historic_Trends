# Historic Stock Trends: How's the Market Really Doing?


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dow_df = pd.read_csv('DowJones-DJIA-100_Year_Historical_Chart.csv')
sp_df = pd.read_csv('S&P_500_Index-90_Year_Historical_Chart.csv')
```

```python
dow_df = dow_df.drop(columns=['series_id', 'nominal'])
```

```python
sp_df = sp_df.drop(columns=['series_id', 'nominal'])
sp_df
```

```python
def get_datetimes(df): #function to convert date to date/time format
    return pd.to_datetime(df.date)
```

```python
dow_df['date'] = get_datetimes(dow_df)
sp_df['date'] = get_datetimes(sp_df)
```

```python
dow_df.set_index('date', inplace=True)
print(dow_df.head())
sp_df.set_index('date', inplace=True)
sp_df
```

```python
dow_ser = pd.Series(dow_df['real'])
sp_ser = pd.Series(sp_df['real'])
```

```python
def test_stationarity(timeseries, window):
    
    #Defining rolling statistics
    rolmean = timeseries.rolling(window=window).mean()
    rolstd = timeseries.rolling(window=window).std()

    #Plot rolling statistics:
    fig = plt.figure(figsize=(12, 8))
    orig = plt.plot(timeseries.iloc[window:], color='#87CEEB',label='Original')
    mean = plt.plot(rolmean, color='#FFC0CB', label='Rolling Mean')
    std = plt.plot(rolstd, color='#008000', label = 'Rolling Std')
    plt.legend(loc='upper left')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
```

```python
test_stationarity(dow_ser, 3)
```

```python
test_stationarity(sp_ser, 3)
```

```python
#importing necessary libraries for statistical modeling
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.arima_model import ARMA,ARMAResults,ARIMA,ARIMAResults
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
decomposition = seasonal_decompose((np.log(dow_ser)), freq=52)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plot statistics
plt.figure(figsize=(12,8))    
plt.subplot(411)
plt.plot(np.log(dow_ser), label='Original', color="blue")
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend', color="blue")
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality', color="blue")
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals', color="blue")
plt.legend(loc='best')
plt.tight_layout()

ts_log_decompose = residual
ts_log_decompose.dropna(inplace=True)

test_stationarity(ts_log_decompose, 10)
```
